```python
import numpy as np
import matplotlib.pyplot as plt
import os, os.path
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
```

### 1. Load the dataset and Visualize Data

Let's start by loading your entire dataset or a sample if it is a larger dataset.

```python
df = pd.read_csv('dataset/dataset.csv')
```

```python
df.head()
```

```python
"""
Iterating throught the dataframe head, and using the filename as path to 
select the image and plotting it amd using the label column as label.
"""

for index, row in df.head().iterrows():
    img = Image.open(row['filename'])
    
    fig, ax = plt.subplots(figsize=(4, 3)) 
    
    ax.imshow(img)
    
    ax.set_title(row['label'])
    ax.axis('off')
    
    plt.show()
```

### 2. Statics of the data. Data set distribution


```python
df.head()
```

```python
# Count occurrences of each label
label_counts = df['label'].value_counts()

# Calculate percentage of each label
label_percentages = label_counts / len(df) * 100

# Plotting
plt.figure(figsize=(12, 6))

# Bar plot
label_counts.plot(kind='bar', color='skyblue')
plt.title('Label Counts')
plt.xlabel('Labels')
plt.ylabel('Count')
plt.xticks(rotation=45)


plt.tight_layout()
plt.show()
```

```python
# Plotting 1 image from each category
plt.figure(figsize=(12, 8))

num_categories = 12
num_cols = 4
num_rows = num_categories // num_cols + (1 if num_categories % num_cols != 0 else 0)

for i, label in enumerate(df['label'].unique(), start=1):
    image_path = df[df['label'] == label]['filename'].iloc[0]
    img = Image.open(image_path)
    plt.subplot(num_rows, num_cols, i)
    plt.imshow(img)
    plt.title(label)
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Get the frequency of each insect
label_counts = df['label'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(label_counts, labels=label_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Insects')
plt.axis('equal') 
plt.show()
```

```python
# Calculate percentages
total_labels = len(df)
percentages = [(count / total_labels) * 100 for count in label_counts]

# Analyzing the percentage of label counts
for label, count, percentage in zip(label_counts.index, label_counts, percentages):
    print(f"{label}: {percentage:.2f}%")

# Calculate average, minimum, and maximum percentage
average_percentage = sum(percentages) / len(percentages)
min_percentage = min(percentages)
max_percentage = max(percentages)

print(f"\nAverage Percentage: {average_percentage:.2f}%")
print(f"Minimum Percentage: {min_percentage:.2f}%")
print(f"Maximum Percentage: {max_percentage:.2f}%")
```
